```python
#import modules
import cv2, time
import matplotlib.pyplot as plt 

# Create a video object
cam = cv2.VideoCapture(0) #zero is used for external camera

frame_list=[]

while cam.isOpened():
    
    #Reading input from webcam
    check_status, frame = cam.read(0)  #video.read() gives a list with two parts: first is status of camera, second is the captured frame

    # Capturing consecutive frames
    frame_list.append(frame)
    
    #Showing the captured frame
    cv2.imshow('Capturing', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break                               #Keep capturing frames and Quit when 'q' is pressed

        
cv2.destroyAllWindows()

plt.imshow(frame_list[0])
plt.show()
```
